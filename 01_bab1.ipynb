```python
# hal 25
import networkx as nx # untuk membuat graf
```

```python
G = nx.Graph()
V = {
    'Jakarta',
    'Bandung',
    'Semarang',
    'Surabaya'
}

E = {
    ('Jakarta', 'Bandung'),
    ('Bandung', 'Semarang'),
    ('Semarang', 'Surabaya'),
    ('Surabaya', 'Jakarta')
}
G.add_nodes_from(V)
G.add_edges_from(E)

nx.draw(G, with_labels=True)
```

```python
print(f"V = {G.nodes}")
print(f"E = {G.edges}")
```

```python
print(f"Graph Order : {G.number_of_nodes()}")
print(f"Graph Size : {G.number_of_edges()}")
```

```python
#Bipartite graph
import pandas as pd # library dataframe,data-data tabular
import numpy as np  # library matematika khususnya buat metode numerik
```

```python
n_nodes = 10
n_edges = 12

bottom_nodes = [ith for ith in range(n_nodes) if ith % 2 == 0]
top_nodes = [ith for ith in range(n_nodes) if ith % 2 == 1]

# jika x adalah suatu bilangan yang akan dibagi
# dam n adalah suatu bilangan yang membagi
# maka x mod n = y, y bagian dari (0 -(n-1)) 

iter_edges = zip(
    np.random.choice(bottom_nodes, n_edges),
    np.random.choice(top_nodes, n_edges)
    
)
edges = pd.DataFrame(
    [{"source":a, "target":b} for a,b in iter_edges]
)

edges


```

```python
edges.values
```

```python
B = nx.Graph()
B.add_nodes_from(bottom_nodes, bipartite=0)
B.add_nodes_from(top_nodes, bipartite=1)
B.add_edges_from([tuple(x) for x in edges.values])
```

```python
from networkx.drawing.layout import bipartite_layout
pos = bipartite_layout(B, bottom_nodes)

nx.draw(B, pos, with_labels=True)
```

```python
#Graph Properties
"""
    1. Integration metric : mengukur sebarapa kuat nodes-nodes saling terkoneksi satu sama lain
    2. Segregration metric : mengukut bagaimana interkoneksi / keterhubungan antara nodes dari suatu grup ke nodes yang mempunyai grup berbeda
    3. Centrality metric : untuk mengukur seberapa suatu nodes itu menjadi pusat, atau berapa dalam sentral /tengah-tengah dari suatu graph
    4. Resilince metric : mengukur seberapa konsisten suatu graph
    
"""

```

```python
#integration metric 
# Distance, path, dan jarak terpendek (shortest path)

#shortest path

path = nx.shortest_path(G, source='Bandung', target='Malang')
print(f"Shortest path from Bandung to malang: {path}")
```

```python
# path ( karakteristik panjang jalur )
# rata-rata semua jalur terpendek diantara pasangan nodes yang mugkin 

nx.average_shortest_path_length(G)
```

```python
#Global dan local efficiency
#global efficiency adalah rata-rata semua jalur terpendek diantara pasangan nodes yang mungkin namun inverse nya
ge = 1/1.4
print(f"Global efficiency: {ge}")

ge_nx = nx.global_efficiency(G)
print(f"Global efficiency: {ge_nx}")
```

```python
#local efficieny 
nx.local_efficiency(G)
```

```python
# Segregation metric
# clustering coefficient

nx.average_clustering(G)
```

```python
nx.clustering(G)
```

```python
#Transitivity metric
#transitivity adalah rasio antara jumlah cluster tertutup (subgraph. 3 nodes dan 2 jalur ) dengan banyaknya kemungkinan cluster tertutup pada graph. 

nx.transitivity(G)
```

```python
#Modularity 
import networkx.algorithms.community as comm

modularity = comm.modularity(G, communities=[['Surabaya', 'Semarang', 'Bandung'], ['Jakarta', 'Malang']])
print(f"Modularity: {modularity}")
```
